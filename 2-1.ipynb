```python
import pandas as pd

df = pd.read_csv('data/fish.csv')

con = df['Species'].isin(['Bream', 'Smelt'])
lw_df = df.loc[con, ['Length2', 'Weight']]
fish_data = lw_df.values.tolist()
fish_target = [1] * 35 + [0] * 14
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

```python
train_input = fish_data[:35]
train_target = fish_target[:35]

test_input = fish_data[35:]
test_target = fish_target[35:]
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
import numpy as np

input_arr = np.array(fish_data)
target_arr = np.array(fish_target)

np.random.seed(42)
index = np.arange(49)
np.random.shuffle(index)
```

```python
input_arr[[1, 3]]
```

```python
train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]
```

```python
test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
```

```python
import matplotlib.pyplot as plt

plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(test_input[:, 0], test_input[:, 1])
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
kn.predict(test_input)
```

## iris

```python
import seaborn as sns
iris = sns.load_dataset('iris')

# setosa, versicolor
# petal_length, petal_width

iris
```

```python
setosa_iris = iris.loc[iris['species'] == 'setosa']
setosa_l = setosa_iris['petal_length'].tolist()
setosa_w = setosa_iris['petal_width'].tolist()

versicolor_iris = iris.loc[iris['species'] == 'versicolor']
versicolor_l = versicolor_iris['petal_length'].tolist()
versicolor_w = versicolor_iris['petal_width'].tolist()
```

```python
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()

length = setosa_l + versicolor_l
width = setosa_w + versicolor_w

iris_data = [[l, w] for l, w in zip(length, width)]
iris_target = [1] * len(setosa_w) + [0] * len(versicolor_w)

kn.fit(iris_data, iris_target)
```

```python
kn.score(iris_data, iris_target)
```

```python
import numpy as np
input_arr = np.array(iris_data)
target_arr = np.array(iris_target)

index = np.arange(100)
np.random.shuffle(index)
```

```python
train_input = input_arr[index[:len(setosa_w)]]
train_target = target_arr[index[:len(setosa_w)]]
test_input = input_arr[index[len(setosa_w):]]
test_target = target_arr[index[len(setosa_w):]]

plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(test_input[:, 0], test_input[:, 1])
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
kn.predict(test_input)
```
